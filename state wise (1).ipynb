###**WHICH PARTY DOMINATED IN NORTHERN STATES**

```python
northern_states = ['up', 'punjab', 'haryana', 'himachal', 'jammukashmir', 'delhi']
df_northern = df_combined[df_combined['State'].isin(northern_states)]
df_north_agg = df_northern.groupby(['State','Party'])['Won'].sum().reset_index()
dominant_parties_north = df_north_agg.loc[df_north_agg.groupby('State')['Won'].idxmax()]
party_wins = dominant_parties_north['Party'].value_counts().reset_index()
party_wins.columns = ['Party', 'Number of Northern States Won']
most_dominant_party = party_wins.loc[party_wins['Number of Northern States Won'].idxmax()]
print("Dominant Party in Each Norhern State :")
for _, row in dominant_parties_north.iterrows():
    print(f"Northern State: {row['State']}")
    print(f"Dominant Party: {row['Party']}, Seats Won: {row['Won']}\n")
print(f"Party with the most wins across Northern Sates: {most_dominant_party['Party']}")
print(f"Number of Northern States Won: {most_dominant_party['Number of Northern States Won']}")
```

###**SEAT DISTRIBUTION BY PARTY IN EACH STATE**

```python
df_combined['Won'] = pd.to_numeric(df_combined['Won'], errors='coerce')
df_agg = df_combined.groupby(['State', 'Party'])['Won'].sum().reset_index()
pivot_table = df_agg.pivot(index='State', columns='Party', values='Won').fillna(0)
state_totals = pivot_table.sum(axis=1).sort_values(ascending=False).index
pivot_table = pivot_table.loc[state_totals]
custom_palette = sns.color_palette("tab20", n_colors=len(pivot_table.columns))
plt.figure(figsize=(30, 30))
sns.set_color_codes('pastel')
pivot_table.plot(kind='bar', stacked=True, figsize=(12, 8), color=custom_palette)
plt.xlabel('State')
plt.ylabel('Seats Won')
plt.title('Seat Distribution by Party in Each State')
plt.legend(title='Party', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```
